```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# MPL работает с типами из numpy
x = np.array([2, 4, 6])
y = np.array([3, 5, 7])

# Построение линейного графика
plt.title("Пример линейного графика")
plt.plot(
    x, y,
    label="Линейная зависимость",
    color="red",
    linestyle='--'
)

plt.legend()                                # отобразить label 
plt.grid()                                  # отобразить сетку

# задаем название осям
plt.xlabel("Ось X")                         
plt.ylabel("Ось Y")

# отобразить график
plt.show()
```

```python
# Более сложный график
days = ['Пн', 'Вт', 'Ср', 'Чт', 'Пт']
sales = np.array([100, 150, 90, 200, 100])

plt.title("Продажи по дням недели")
plt.plot(
    days, sales,
    label='Продажи',
    color='indigo',
    marker='o',
)
plt.legend()

plt.show()
```

```python
plt.title('Гистограмма нормального распределения')

data = np.random.normal(170, 30, 250)

plt.hist(
    data,
    color='orange',
    edgecolor='black'
)
plt.grid()
plt.ylabel('Количество элементов')

plt.show()
```
